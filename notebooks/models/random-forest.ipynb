```python
df_eng = pd.read_csv('/Users/yashwanthkaruparthi/Developer/energy_demand/research/data/8-feat-eng/feat-lag.csv')
```

```python
df_eng.drop(columns=['Unnamed: 0', 'index'], inplace = True)
```

```python
df_eng = df_eng.dropna()
```

```python
# df_eng['date'] = pd.to_datetime(df_eng[['year', 'month', 'day']])
```

```python
# print(df_eng[['year', 'month', 'day', 'date']].head())
```

```python
X = df_eng.loc[:, df_eng.columns != 'daily_demand'].values
y = df_eng.loc[:, 'daily_demand'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# X_train = pd.DataFrame(X_train, columns=X.columns)
# X_test = pd.DataFrame(X_test, columns=X.columns)
```

```python
print(y_test)
```

```python
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
rf_regressor.fit(X_train, y_train)
# rf_regressor.fit(X_train.drop(columns=['date', 'year', 'month', 'day']), y_train)


```

```python
y_pred = rf_regressor.predict(X_test)

# y_pred = rf_regressor.predict(X_test.drop(columns=['date', 'year', 'month', 'day']))
```

```python
r_sq = metrics.r2_score(y_test,y_pred)
print(f"R^2 Score: {r_sq}")
```
